```python
#load in the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
```

```python
#import the dataframe
df = pd.read_csv('sports_survey_2021.csv')
df.head
```

```python
df.info()
```

```python
df.describe()
```

```python
#look at gender
gender_counts = df['S1'].value_counts() 
  
#count plot
plt.figure(figsize=(8, 6)) 
plt.bar(gender_counts.index, gender_counts, color='blue') 
plt.title('Count Plot of Gender') 
plt.xlabel('Gender') 
plt.ylabel('Count') 
plt.show()
#1 is Male, 2 is Female
#Definitiely more men
```

```python
#look at age
age_counts = df['S2'].value_counts() 
  
#count plot
plt.figure(figsize=(8, 6)) 
plt.bar(age_counts.index, age_counts, color='blue') 
plt.title('Count Plot of Age') 
plt.xlabel('Age') 
plt.ylabel('Count') 
plt.show()
```

```python
#age again
plt.figure(figsize=(8, 6))
sns.histplot(df['S2'])
plt.title('Distribution of Age Variable')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#look at household income
hhincome_counts = df['D4'].value_counts() 
  
#count plot
plt.figure(figsize=(8, 6)) 
plt.bar(hhincome_counts.index, hhincome_counts, color='blue') 
plt.title('Count Plot HH Income') 
plt.xlabel('HH income') 
plt.ylabel('Count') 
plt.show()
#1: under 25,000
#2: 25,000-49,999
#3 50,000-74,999
#4 75,000-99,999
#5 100,000-149,999
#6 150,000 or more
```

```python
#look at Employment
employment_counts = df['D5'].value_counts() 
  
#count plot
plt.figure(figsize=(8, 6)) 
plt.bar(employment_counts.index, employment_counts, color='blue') 
plt.title('Count Plot of Employment') 
plt.xlabel('Employment') 
plt.ylabel('Count') 
plt.show()
# 1: Full time, 2: Part time, 3: Not Employed, 4: Homemaker, 5: Retired, 6: Prefer to not say
```

```python
#look at Educational Attainment
education_counts = df['D6'].value_counts() 
  
#count plot
plt.figure(figsize=(8, 6)) 
plt.bar(education_counts.index, education_counts, color='lightgreen') 
plt.title('Count Plot of Education') 
plt.xlabel('Education') 
plt.ylabel('Count') 
plt.show()
# 1: Some high school
#2: High school grad
#3: Vocational/trade school
#4: Some college
#5: College grad
#6: Post-graduate degree
#7: prefer not to say
```

```python
#look at Hours Watching Sports
hourswatched_counts = df['S12r3'].value_counts() 
  
#count plot
plt.figure(figsize=(8, 6)) 
plt.bar(hourswatched_counts.index, hourswatched_counts, color='lightgreen') 
plt.title('Count Plot of Hours of Sports Watched') 
plt.xlabel('Hours') 
plt.ylabel('Count') 
plt.show()
#Definitely skewed towards 0
```

```python
#look at Self-Reported Fandom: NFL
fannfl_counts = df['S15r1'].value_counts() 
  
#count plot
plt.figure(figsize=(8, 6)) 
plt.bar(fannfl_counts.index, fannfl_counts, color='lightgreen') 
plt.title('Count Plot of Self-Reported NFL Fandom') 
plt.xlabel('Fan Rank') 
plt.ylabel('Count') 
plt.show()
#Definitely skewed towards 5 and 6 (1: Not a fan ; 6: Obsessed Fan of this Sport)
```
