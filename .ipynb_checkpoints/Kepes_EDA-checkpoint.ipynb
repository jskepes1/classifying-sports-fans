```python
#load in the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
```

```python
#import the dataframe
df = pd.read_csv('sports_survey_2021.csv')
```

```python
#clean some of the data
#clean some of the SES Data
df = df[df['D5'] != 6]
df = df[df['D6'] != 7]
df.describe

columns_to_fill = ['S15r1', 'S15r2', 'S15r3', 'S15r4', 'S15r5', 'S15r6', 'S15r7', 'S15r8', 'S15r9', 'S15r10', 'S15r11']

for col in columns_to_fill:
    df[col] = df[col].fillna(0)
```

```python
#create an index of the 11 sports
df['averageS15'] = df[['S15r1', 'S15r2', 'S15r3', 'S15r4', 'S15r5', 'S15r6', 'S15r7', 'S15r8', 'S15r9', 'S15r10', 'S15r11']].mean(axis=1)
df = df[df['averageS15'] != 0]
df.averageS15.head()
```

```python
ses_columns = ['S1', 'S2', 'D4', 'D5', 'D6']
df_ses = df[ses_columns]
```

```python
#Create an Index of Socioeconomic Status?
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(df_ses)
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out())
# Dimensionality Reduction with PCA
pca = PCA(n_components=1)
socioeconomic_index = pca.fit_transform(encoded_df)
df['ses_index'] = socioeconomic_index
```

```python
df['ses_index'] = socioeconomic_index
print(df.ses_index.describe())
```

```python
#look at SES
ses_counts = df['ses_index'].value_counts() 
  
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['ses_index'], kde=True)
plt.xlabel('Socioeconomic Index')
plt.ylabel('Frequency')
plt.title('Distribution of Socioeconomic Index')
plt.show()
#definitely skewed right for some reason with a bimodal distribution
```

```python
# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='ses_index', y='averageS15')

# Adding labels and title
plt.xlabel('SES')
plt.ylabel('Fandom Average')
plt.title('Scatter Plot')

# Show plot
plt.show()
```
